```python
import subprocess
import sys
import os

def install_if_missing(package):
    """Check if package is installed, if not install it."""
    try:
        __import__(package)
    except ImportError:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])

# Install packages
install_if_missing('pandas')
install_if_missing('numpy')
install_if_missing('matplotlib')
install_if_missing('pylab')
install_if_missing('IPython')

# Now import them
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

import time
import math

import pylab as py
from IPython.display import HTML
```

```python
# For ISP information see actuator summary document in the shared drive
I_sp_cold_gas = 72  # seconds
I_sp_FEEP = 2500  # seconds
I_sp_MEP = 850  # seconds

M_spacecraft = 500  # kg

# For total period impulse calculations see perturbed RK4 simulation in the shared drive
I_total_period_max = 2*4.28e+00 #Ns
g_0 = 9.81  # m/s^2
```

```python
# The calculation is based on the Wikipedia site https://en.wikipedia.org/wiki/Specific_impulse#Specific_impulse_in_seconds
M_fuel_cold_gas = I_total_period_max / (I_sp_cold_gas * g_0)
M_fuel_FEEP = I_total_period_max / (I_sp_FEEP * g_0)
M_fuel_MEP = I_total_period_max / (I_sp_MEP * g_0)

print(f"Fuel mass for cold gas thrusters: {M_fuel_cold_gas*1000:.2f} g")
print(f"Fuel mass for FEEP thrusters: {M_fuel_FEEP*1e6:.2f} mg")
print(f"Fuel mass for MEP thrusters: {M_fuel_MEP*1e6:.2f} mg")
```

```python
T_missio_lifetime_years_2 = 2
T_mission_lifetime_seconds_2 = T_missio_lifetime_years_2 * 365.25 * 24 * 3600

# From assuming a circular orbit at 100e6m
T_period_seconds = 314010.15

N_periods_2 = T_mission_lifetime_seconds_2 / T_period_seconds

T_missio_lifetime_years_5 = 5
T_mission_lifetime_seconds_5 = T_missio_lifetime_years_5 * 365.25 * 24 * 3600

# From assuming a circular orbit at 100e6m
T_period_seconds = 314010.15

N_periods_5 = T_mission_lifetime_seconds_5 / T_period_seconds
```

```python
I_total_mission_2 = I_total_period_max * N_periods_2
I_total_mission_5 = I_total_period_max * N_periods_5
print(f"Total impulse over 2 years mission lifetime: {I_total_mission_2:.2f} Ns")
print(f"Total impulse over 5 years mission lifetime: {I_total_mission_5:.2f} Ns")
```

```python
M_fuel_cold_gas_total = M_fuel_cold_gas * N_periods_2
M_fuel_FEEP_total = M_fuel_FEEP * N_periods_5
M_fuel_MEP_total = M_fuel_MEP * N_periods_5

print(f"Total fuel mass for cold gas thrusters over 2 years: {M_fuel_cold_gas_total:.2f} kg")
print(f"Total fuel mass for FEEP thrusters over 5 years: {M_fuel_FEEP_total:.2f} kg")
print(f"Total fuel mass for MEP thrusters over 5 years: {M_fuel_MEP_total:.2f} kg")
```
